```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('covid_19_indonesia_time_series_all.csv')
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'], format="%m/%d/%Y")
df['Case Fatality Rate'] = df['Case Fatality Rate'].str.rstrip('%').astype('float') / 100.0
df['Case Recovered Rate'] = df['Case Recovered Rate'].str.rstrip('%').astype('float') / 100.0
df.head()
```

```python
df.corr()
```

```python
mask = np.triu(df.corr())
plt.style.use('seaborn')
fg, ax = plt.subplots(figsize=(18, 7), dpi=200)
sns.heatmap(df.corr(), mask=mask, cmap='Reds', annot=True, linewidths=1)
plt.title('KORELASI TIAP FEATURE', fontweight='bold', fontsize=20)
plt.xticks(rotation=30)
plt.show()
```
